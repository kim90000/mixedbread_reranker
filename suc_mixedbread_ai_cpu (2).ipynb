```python
from mxbai_rerank import MxbaiRerankV2
import PyPDF2

model = MxbaiRerankV2("mixedbread-ai/mxbai-rerank-large-v2")

query = "What is the topic of the book?"
pdf_file_path = "/content/sample.pdf"  # استبدل بمسار ملف PDF الخاص بك

documents = []
try:
    with open(pdf_file_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)  # استخدم PdfReader بدلاً من PdfFileReader

        for page_num in range(len(pdf_reader.pages)): # استخدم pages بدلاً من numPages و len() للحصول على عدد الصفحات
            page = pdf_reader.pages[page_num] # استخدم pages كقائمة بدلاً من getPage
            page_text = page.extract_text()
            if page_text.strip():  # تأكد من أن الصفحة ليست فارغة بعد إزالة المسافات البيضاء
                documents.append(page_text)

except FileNotFoundError:
    print(f"Error: PDF file not found at path: {pdf_file_path}")
    exit()
except Exception as e:
    print(f"Error reading PDF file: {e}")
    exit()


# Lets get the scores
results = model.rank(query, documents, return_documents=True, top_k=3)

print(results)
```

```python
from mxbai_rerank import MxbaiRerankV2
import PyPDF2

model = MxbaiRerankV2("mixedbread-ai/mxbai-rerank-large-v2")

def extract_answer_from_pdf(pdf_file_path, query):
    documents = []
    try:
        with open(pdf_file_path, 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            num_pages = len(pdf_reader.pages) # للحصول على عدد الصفحات مقدما

            if query.lower().startswith("كم عدد صفحات"): # التعامل مع سؤال عدد الصفحات مباشرة
                return f"{num_pages} صفحة"

            for page_num in range(num_pages):
                page = pdf_reader.pages[page_num]
                page_text = page.extract_text()
                if page_text.strip():
                    documents.append(page_text)

    except FileNotFoundError:
        return "Error: PDF file not found."
    except Exception as e:
        return f"Error reading PDF file: {e}"

    if not documents: # حالة الملف الفارغ او بدون نصوص
        return "لم يتم العثور على نصوص في ملف PDF."


    results = model.rank(query, documents, return_documents=True, top_k=1) # الحصول على أفضل وثيقة فقط

    if not results: # حالة عدم وجود نتائج ترتيب (غير محتمل ولكن للتحوط)
        return "لم يتم العثور على إجابة."

    top_document = results[0].document # الوثيقة الأعلى ترتيبا

    if query.lower().startswith("من مؤلف"): # التعامل مع سؤال المؤلف بشكل مبسط
        author_patterns = ["تأليف ", "كتب بواسطة ", "المؤلف هو ", "مؤلف الكتاب هو ", "بواسطة "] # أنماط للبحث عن المؤلف
        for pattern in author_patterns:
            start_index = top_document.lower().find(pattern.lower())
            if start_index != -1:
                author_name_start = start_index + len(pattern)
                # محاولة استخراج الاسم (بشكل بسيط، قد يحتاج إلى تحسين)
                author_name_end = top_document.find(".", author_name_start) # البحث عن نقطة نهاية الجملة
                if author_name_end == -1:
                    author_name_end = len(top_document) # إذا لم توجد نقطة، نأخذ حتى نهاية الوثيقة
                author_name = top_document[author_name_start:author_name_end].strip()
                return author_name.strip() # إرجاع الاسم بعد إزالة المسافات الزائدة

    # إذا لم يتم العثور على إجابة محددة أو كان السؤال نوعًا آخر، يتم إرجاع الوثيقة الأعلى ترتيبًا كمقتطف (يمكن تعديل هذا)
    return "لم يتم العثور على إجابة محددة. الوثيقة الأكثر صلة:\n" + top_document[:500] + "..." # إرجاع جزء من الوثيقة إذا لم يتم العثور على إجابة محددة


# --- مثال الاستخدام ---
pdf_file_path = "your_pdf_file.pdf" # استبدل بمسار ملف PDF الخاص بك
query_author = "من مؤلف الكتاب؟"
query_pages = "كم عدد صفحات الكتاب؟"
query_other = "ما هي المواضيع الرئيسية في الكتاب؟" # سؤال آخر غير محدد التعامل معه بعد

answer_author = extract_answer_from_pdf(pdf_file_path, query_author)
answer_pages = extract_answer_from_pdf(pdf_file_path, query_pages)
answer_other = extract_answer_from_pdf(pdf_file_path, query_other)


print(f"السؤال: {query_author}\nالإجابة: {answer_author}\n")
print(f"السؤال: {query_pages}\nالإجابة: {answer_pages}\n")
print(f"السؤال: {query_other}\nالإجابة: {answer_other}\n")
```
